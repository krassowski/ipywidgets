```python
import threading

import ipywidgets as widgets

button = widgets.Button(description='Start background thread')
out = widgets.Output()

def on_click(_):
    def write_to_output():
        with out:
            print('captured from background thread')

    thread = threading.Thread(target=write_to_output)
    thread.start()
    thread.join()

button.on_click(on_click)
widgets.VBox([button, out])
```
